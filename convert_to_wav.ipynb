```python
from google.colab import files
uploaded = files.upload()
```

```python
from pydub import AudioSegment
import os

# Buat folder output
output_folder = "converted_wav"
os.makedirs(output_folder, exist_ok=True)

# Loop semua file yang diupload
for filename in uploaded.keys():
    if filename.endswith(".ogg"):
        wav_filename = os.path.splitext(filename)[0] + ".wav"
        wav_path = os.path.join(output_folder, wav_filename)

        try:
            sound = AudioSegment.from_ogg(filename)
            sound.export(wav_path, format="wav")
            print(f"✔️ {filename} -> {wav_filename}")
        except Exception as e:
            print(f"❌ Gagal konversi {filename}: {e}")
```

```python
import zipfile

# Buat ZIP
zipf = zipfile.ZipFile("converted_wav.zip", "w")
for root, dirs, files in os.walk("converted_wav"):
    for file in files:
        zipf.write(os.path.join(root, file))
zipf.close()

# Download hasil ZIP
from google.colab import files as colab_files
colab_files.download("converted_wav.zip")
```
